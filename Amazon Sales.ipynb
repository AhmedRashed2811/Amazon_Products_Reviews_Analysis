# Data Preprocessing


Dropping Unnecessary Columns "Won't Be Used In Analysis"

```python
df.drop(['img_link', 'product_link','discounted_price','about_product',"user_name","review_id"], axis=1, inplace = True)
```

Getting Number Of Users Who Gave Review On Each Product

```python
df["number_users"] = df["user_id"].apply(lambda x: x.count(',')+1)
```

Creating a New Dataframe For Preprocessing and With no Duplicates

```python
df_Preprocessing = df.drop_duplicates()
```

Getting a List of All Attributes in the Dataframe

```python
columns_1=df_Preprocessing.columns.to_list()
columns_1.remove('rating_count')
```

Creating a Function that will Delete Unnecessary Characters "," & "₹"

```python
def change_type(x):
  if type(x) == float:
    return x
  else:
    return x.replace(",","").replace("₹","")
```

Converting Attributes to Suitable Types and Dealing With Missing Values

```python
#Changing "rating_count" attribute to float and fill NaN

df_Preprocessing["rating_count"] = df_Preprocessing["rating_count"].apply(lambda x: change_type(x))
df_Preprocessing["rating_count"]  = pd.to_numeric(df_Preprocessing["rating_count"])
df_Preprocessing["rating_count"] = df_Preprocessing["rating_count"].fillna(1)


#Changing "actual_price" attribute to float and fill NaN

df_Preprocessing["actual_price"] = df_Preprocessing["actual_price"].apply(lambda x: change_type(x))
df_Preprocessing["actual_price"]  = pd.to_numeric(df_Preprocessing["actual_price"])


#Getting List of wrong ratings "Dupliacated Ones"
list_of_wrong_ratings = df_Preprocessing[df_Preprocessing.duplicated(columns_1)]["rating_count"].to_list()

#Dealing with the wrong ratings
new_ratings = []

for i in df_Preprocessing["rating_count"]:
    
    #If wrong rating
  if i in list_of_wrong_ratings:
    new_ratings.append(i+1)
    
    #Else
  else:
    new_ratings.append(i)

df_Preprocessing["ratings_count"] = new_ratings
```

Take a Look On the Data

```python
df_Preprocessing.head(3)
```

Dropping Unnecessary Columns "Won't Be Used In Analysis" and Dropping Duplicated Products

```python
df_Preprocessing.drop(["rating_count", "user_id"] , axis = 1, inplace = True)
df_Preprocessing = df_Preprocessing.drop_duplicates("product_id")
```

Take a Look On the Data

```python
df=df_Preprocessing
df_Preprocessing.head(3)
```

Splitting Categories and Creating an Attribute for Each One

```python
#Getting the number of categories
number_of_categories = len(df["category"].str.split('|',expand=True).columns)

#Splitting Categories and dropping "category" unnecessary attribute
df = pd.concat([df, df["category"].str.split('|',expand=True)], axis=1).drop("category", axis = 1)

#Renaming categories new attributes
df.rename(columns = {0:'cat1', 1:'cat2',2:'cat3', 3:'cat4', 4:'cat5', 5:'cat6', 6:'cat7', 7:'cat8',8:'cat9'}, inplace = True)

#Filling NaNs with empty string
df.fillna("", inplace =True)

#Take a look on the data
df.head(3)
```

There is one record with missing rating, so Filling it With 3.5 as Average, Then Converting the ratings to Float

```python
df.loc[df["rating"] == '|', "rating"] = 3.5
df["rating"] = pd.to_numeric(df["rating"])
df.head(3)
```

# Analysis Functions

Getting Average Rating Per Category

```python
def get_category_avg_rating(cat):
  sum_of_ratings = 0
  counter = 0

  for  i in range(0,df.shape[0]):
    if cat in df.iloc[i].values:
      sum_of_ratings+= df.iloc[i]["rating"]
      counter += 1

  if counter == 0: return "Category Not Found"
  else: return sum_of_ratings/counter
```

```python
get_category_avg_rating("Computers&Accessories")
```

TopN Function by Criteria with Sorting

```python
def topN(category, n, criteria, sortby = "rating"):
    
  if category == None or type(category) != str:
    return "Must Enter Category..."

  if n < 1 or type(n) != int:
    return f"{n} Must be a Positive Integer Number..."

  if criteria not in df.columns:
    return f"Criteria Not Found...Please Choose From {df.columns.values.tolist()}" 
  
  if sortby not in df.columns:
    return f"Sortby Item Not Found...Please Choose From {df.columns.values.tolist()}" 
    
  x = pd.DataFrame()
  temp = 0
  for i in range(1,number_of_categories):
    if df[df[f"cat{i}"] == category].empty == False and temp == 0:
      x = df[df[f"cat{i}"] == category]
      temp += 1

    elif df[df[f"cat{i}"] == category].empty == True or temp != 0:
      x = pd.concat([x, (df[df[f"cat{i}"] == category])])
    
  if df[sortby].dtype == float:
      return x.nlargest(n, criteria)[["product_name", "actual_price","rating","ratings_count"]].sort_values(by=[sortby], ascending=False)
  else:
      return x.nlargest(n, criteria)[["product_name", "actual_price","rating","ratings_count"]].sort_values(by=[sortby])
    
```
